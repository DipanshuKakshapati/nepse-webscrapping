```python
root = html.fromstring(data)
print(html.tostring(root, pretty_print=True).decode('utf-8'))
```

```python
# finding all <tr> elements in the HTML
rows = root.xpath('//tr')
```

### defining empty lists to store data

```python
Sn = []
Date = []
Close = []
High = []
Low = []
Absolute_Change = []
Percentage_Change = []
Fifty_Two_Weeks_High = []
Fifty_Two_Weeks_Low = []
Turnover_Values = []
Turnover_Volume = []
Total_Transaction = []
```

### getting individual cell values of individual rows

```python
# iterating over all the individual rows
for row in rows:
    # finding all the <td> tags within this row. Here, <td> tags represent the individual values/cells of individual rows.
    cells = [td.text_content().strip() for td in row.xpath('.//td')]
    if len(cells) == 12:  # ensuring there are exactly 9 columns, as expected.
        Sn.append(cells[0])
        Date.append(cells[1])
        Close.append(cells[2])
        High.append(cells[3])
        Low.append(cells[4])
        Absolute_Change.append(cells[5])
        Percentage_Change.append(cells[6])
        Fifty_Two_Weeks_High.append(cells[7])
        Fifty_Two_Weeks_Low.append(cells[8])
        Turnover_Values.append(cells[9])
        Turnover_Volume.append(cells[10])
        Total_Transaction.append(cells[11])
```

### creating a dataframe to store all data

```python
# creating a DataFrame from the webscrapped data
df = pd.DataFrame({
    'Sn': Sn,
    'Date': Date,
    'Close': Close,
    'High': High,
    'Low': Low,
    'Absolute_Change': Absolute_Change,
    'Percentage_Change': Percentage_Change,
    'Fifty_Two_Weeks_High': Fifty_Two_Weeks_High,
    'Fifty_Two_Weeks_Low': Fifty_Two_Weeks_Low,
    'Turnover_Values': Turnover_Values,
    'Turnover_Volume': Turnover_Volume,
    'Total_Transaction': Total_Transaction
})
```

### formatting data types of all columns in the required format

```python
df['Sn'] = df['Sn'].astype(int)
df['Close'] = df['Close'].str.replace(',', '').astype(float)
df['High'] = df['High'].str.replace(',', '').astype(float)
df['Low'] = df['Low'].str.replace(',', '').astype(float)
df['Absolute_Change'] = df['Absolute_Change'].str.replace('[^0-9.-]', '', regex=True).astype(float)
df['Percentage_Change'] = df['Percentage_Change'].str.replace('[()%]', '', regex=True).str.replace('%', '').astype(float)
df['Fifty_Two_Weeks_High'] = df['Fifty_Two_Weeks_High'].str.replace(',', '').astype(float)
df['Fifty_Two_Weeks_Low'] = df['Fifty_Two_Weeks_Low'].str.replace(',', '').astype(float)
df['Turnover_Values'] = df['Turnover_Values'].str.replace(',', '').astype(float)
df['Turnover_Volume'] = df['Turnover_Volume'].str.replace(',', '').astype(float)
df['Total_Transaction'] = df['Total_Transaction'].str.replace(',', '').astype(float)
```

```python
# query to create a table in the database
cursor.execute('''
CREATE TABLE nepse (
    SN INT IDENTITY(1,1), 
    Date DATE,
    [Close] REAL, 
    [High] REAL, 
    [Low] REAL,
    Absolute_Change REAL,
    Percentage_Change REAL,
    Fifty_Two_Weeks_High REAL,
    Fifty_Two_Weeks_Low REAL,
    Turnover_Values REAL,
    Turnover_Volume REAL,
    Total_Transaction REAL,
    PRIMARY KEY (SN)
)
''')
```

### inserting the data from the dataframe into the newly created table

```python
data_to_insert = df.copy() 

for _, row in data_to_insert.iterrows():
    insert_query = '''
        INSERT INTO nepse (Date, [Close], [High], [Low], Absolute_Change, Percentage_Change, Fifty_Two_Weeks_High, Fifty_Two_Weeks_Low, Turnover_Values, Turnover_Volume, Total_Transaction)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    '''
    params = (row['Date'], row['Close'], row['High'], row['Low'], row['Absolute_Change'], row['Percentage_Change'], row['Fifty_Two_Weeks_High'], row['Fifty_Two_Weeks_Low'], row['Turnover_Values'], row['Turnover_Volume'], row['Total_Transaction'])
    cursor.execute(insert_query, params)
```
